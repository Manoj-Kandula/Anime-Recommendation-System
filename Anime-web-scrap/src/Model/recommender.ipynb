# Modelling

```python
from sklearn.preprocessing import LabelEncoder,MultiLabelBinarizer,StandardScaler
from sklearn.neighbors import NearestNeighbors
```

```python
le = LabelEncoder()
data1['type'] = le.fit_transform(data1['type'])
mlb = MultiLabelBinarizer()
genre = mlb.fit_transform(data1['genre'].str.split(','))
genre_df = pd.DataFrame(genre,columns=mlb.classes_)
data1_copy = data1.copy()
data1_copy = data1_copy.drop(columns=['genre'])
data2 = pd.concat([data1,genre_df],axis=1)
features = data2[['type', 'episodes', 'rating', 'members'] + list(genre_df.columns)]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
print(features.isna().sum())
features.fillna(features.mean(), inplace=True)
```

```python
print(features.isna().sum())
```

```python
features_scaled = scaler.fit_transform(features)
print(features[features.isna().any(axis=1)])
```

```python
knn = NearestNeighbors(n_neighbors=6, metric='euclidean')
knn.fit(features_scaled)
```

```python
def pred(anime_name = 'Cowboy Bebop'):
    anime_id = data1[data1['name'] == anime_name].iloc[0,0]
    anime_features = features_scaled[anime_id - 1].reshape(1, -1)
    distances, indices = knn.kneighbors(anime_features)
    similar_anime_ids = indices.flatten()[1:]
    similar_anime = data1.loc[similar_anime_ids, 'name']
    print(f"Recommended anime similar to {data1.loc[anime_id - 1, 'name']}:")
    print(similar_anime)
pred()
```

```python
pred('Cowboy Bebop')
```

```python
data1 = data1.iloc[:features_scaled.shape[0], :]
features_scaled = features_scaled[:data1.shape[0], :]
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA(n_components=2)
features_2d = pca.fit_transform(features_scaled)
data1['PCA1'] = features_2d[:, 0]
data1['PCA2'] = features_2d[:, 1]
```

```python
def plot_knn_recommendations(anime_id):
    anime_features = features_scaled[anime_id - 1].reshape(1, -1)
    distances, indices = knn.kneighbors(anime_features)
    similar_anime_ids = indices.flatten()
    plt.figure(figsize=(10, 6))
    plt.scatter(data1['PCA1'], data1['PCA2'], alpha=0.5, label="All Anime")
    plt.scatter(data1.loc[anime_id - 1, 'PCA1'], data1.loc[anime_id - 1, 'PCA2'], 
                color='red', s=100, edgecolors='black', label="Selected Anime")
    for idx in similar_anime_ids[1:]:
        plt.scatter(data1.loc[idx, 'PCA1'], data1.loc[idx, 'PCA2'], 
                    color='green', s=100, edgecolors='black')
    plt.xlabel("PCA 1")
    plt.ylabel("PCA 2")
    plt.title(f"Anime Recommendations for {data1.loc[anime_id - 1, 'name']}")
    plt.legend()
    plt.show()
plot_knn_recommendations(anime_id=21)
```
